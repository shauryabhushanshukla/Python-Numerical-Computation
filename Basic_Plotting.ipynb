## Plotting Voltage vs. Time (Continuous Lines)
Goal: Visualize a mathematical function like a sine wave: $V(t) = \sin(t)$.

To do this, you need two arrays: one for the x-axis (Time) and one for the y-axis (Voltage).

np.linspace: Creates a smooth list of time points.

plt.plot(x, y): Connects these points with a line.

Labels: Always label axes with units!

```python
# 1. Generate Data
t = np.linspace(0, 10, 100)  # 100 points from 0 to 10 seconds
V = np.sin(t)                # Voltage formula

#2. Create Plot
plt.figure()                 # Create a new blank figure
plt.plot(t, V)               # Plot t on x-axis, V on y-axis


# 3. Add Labels (Crucial for Science)
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("Voltage vs. Time Signal")

# 4. Show it
plt.show()
```

Overlays: Theory vs. Experiment
Goal: Compare a smooth theoretical curve against discrete, noisy experimental data points.

Theory: Use a high density of points (e.g., 100 points) and a line style (-).

Experiment: Use fewer points (e.g., 10 points) and markers (o).

label=: Assigns a name to each dataset for the legend.

```python
import matplotlib.pyplot as plt
import numpy as np

import unicodedata as ud   # 

A = 1
T = 2* np.pi
omega = (2 *np.pi) / T
t =np.linspace(0,10,100)
Y = A * np.sin(omega *t)

#plotting
plt.figure()
plt.plot(t,Y)

#labelling
plt.xlabel("Time (s)")
plt.ylabel("Displacement (m)")
plt.title(" Wave on string")

plt.show()
```

##  Plotting a line 

```python
X = [0,1,3] # point-1 = (0,1)
Y = [0,2,-1] # point-1 = (0,2)

plt.plot(X,Y, 'ms--', label = 'line')  # m for magenta, s for square, --- for dotted  
plt.legend()

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# writing a letter X

plt.plot([0,1],[1,0], 'm', label = 'magenta line', linewidth= 5)
plt.plot([1,0],[1,0],'r', label =  'red line', linewidth = 2.5)
plt.legend()
plt.show()
```
