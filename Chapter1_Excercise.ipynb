Q-6) Compute the surface area and volume of a cylinder with radius 5 and height 3.

```python
# Cylinder
r = 5
h = 3

# Surface area = 2*pi*r*h + pi*r**2
Surface_area = 2* math.pi * r * h + math.pi * r**2
volume = math.pi * r**2 * h
print(f"Surface Area of Cylinder is {Surface_area} unitsq\nVolume of Cylinder is {volume} unitsq")
```
