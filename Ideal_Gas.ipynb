```python
import numpy as np
import matplotlib.pyplot as plt  # The standard import

# 1. Generate Data (Using NumPy)
V = np.linspace(0.5, 10, 50)  # Volume: 0.5 to 10 m^3 (50 points)
n = 2.0        # moles
R = 8.314      # J/(mol K)
T = 300.0      # Kelvin
P = (n * R * T) / V  # Vectorized Calculation

# 2. Create the Figure
# figsize=(width, height) in inches
plt.figure(figsize=(8, 5)) 

# 3. Plot the Data
# V is x-axis, P is y-axis
# '-o' means "solid line with circle markers"
# color can be 'red', 'blue', 'k' (black), or hex codes
plt.plot(V, P, '-o', color='crimson', label='Isotherm (300K)')

# 4. Add Labels (Mandatory for Science!)
plt.title(f'Ideal Gas Law: Pressure vs Volume (n={n}mol)')
plt.xlabel('Volume ($m^3$)')
plt.ylabel('Pressure (Pa)')

# 5. Final Touches
plt.grid(True, linestyle='--', alpha=0.7)  # Add a grid
plt.legend()  # Show the legend (uses the 'label' from step 3)

plt.show()
```
