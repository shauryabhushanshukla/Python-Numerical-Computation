A leap year is a year that has an extra day, February 29, to keep the timescale on track. Leap years occur on years that are exactly divisible by 4, unless it is exactly divisible by 100, unless it is divisible by 400. For example, the year 2004 is a leap year, the year 1900 is not a leap year, and the year 2000 is a leap year.
Compute the number of leap years between the years 1500 and 2010.

```python
start_year = 1500
end_year = 2010

# Count multiples of 4
year_d_4 = (end_year // 4) - ((start_year - 1) // 4)

# Count multiples of 100
year_d_100 = (end_year // 100) - ((start_year - 1) // 100)

# Count multiples of 400
year_d_400 = (end_year // 400) - ((start_year - 1) // 400)

# Total leap years
total_year = year_d_4 - year_d_100 + year_d_400

print(f"Total count of leap years between {start_year} to {end_year} is {total_year}")
```
